```python
X = df.iloc[:,0:13]
X.head(2)
```

```python
### Frist do train and test split 

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split (X,Y,test_size=0.33,random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
Parameter={'C':[0.5,1,3,5,8,10,20,50,30,31,45,50],'penalty':['l1','l2','elesticnet']}
```

```python
### using logestic regression
from sklearn.linear_model import LogisticRegression
Regr = LogisticRegression()
log_reg=GridSearchCV(Regr,Parameter,cv=5,scoring ='accuracy')
log_reg.fit(x_train, y_train)
```

```python
print(log_reg.best_estimator_)
```

```python
## cheking the which parameter has selected 
print(log_reg.best_params_)
```

```python
## predecting 
predction_logestic=log_reg.predict(x_test)
```

```python
## finding the score 
score = log_reg.score(x_test, y_test)
print(score)
```

```python
### confussion matrix for logestic 
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, predction_logestic)
print(cm)
```

```python
###Deision tree 

from sklearn.tree import DecisionTreeRegressor

treereg=DecisionTreeRegressor()
treereg.fit(x_train,y_train)
```
